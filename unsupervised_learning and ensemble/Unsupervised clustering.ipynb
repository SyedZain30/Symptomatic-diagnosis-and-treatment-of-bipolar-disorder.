```python
import matplotlib.pyplot as plt
```

```python
import random
random.seed(0)
```

```python
bp1=[]
bp2=[]
bp3=[]
for i in range(0,200):
    bp1.append(random.randint(0,10))
    bp2.append(random.randint(25,35))
    bp3.append(random.randint(50,60))
```

```python
y1=[]
y2=[]
y3=[]
for i in range(0,200):
    y1.append(random.uniform(0,5))
    y2.append(random.uniform(0,5))
    y3.append(random.uniform(0,5))
```

```python
plt.figure(figsize=(15,6))
plt.scatter(bp1,y1)
plt.scatter(bp2,y2)
plt.scatter(bp3,y3)
plt.show()
```
